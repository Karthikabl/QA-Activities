```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
tdata=pd.read_excel("C:/Users/user/Desktop/Data Science/ACTIVITY/Dataset_w6.xlsx")
tdata.head()
```

```python
# 1. Find out how many countries are there in the dataset.
Nc = tdata[0:]
NC= len(Nc)
#Nc=tdata.country.count()
print('Number of countries in the dataset :', NC)
```

```python
# 2. Examine the data by using visualizations.
plt.figure(figsize=(10,6))
sns.scatterplot(data=tdata,x='GDP per capita (current US$)',y='Population in thousands (2017)', color='teal').set(title='Population vs GDP Per Capita')
plt.xlabel('GDP per capita (current US$)')
plt.ylabel('Population in thousands (2017)')
plt.show()
```

```python
fig=plt.subplots(figsize=(15,5))
plt.xticks(rotation=90)
sns.barplot(x='Region',y='Population in thousands (2017)', data=tdata).set(title='Population in different Regions')
```

```python
# 3. Find out how many countries with respect to the region are available in the dataset
#data1= tdata.groupby('Region')['country'].count().sort_values()
#data1
data1 = tdata['Region'].value_counts()
print('The countries with respect to the regions available in the dataset are : \n',data1)
```

```python
# 4.On average which region has the highest

# a. GDP per capita (current US$)
data2= tdata.groupby('Region')['GDP per capita (current US$)'].mean().round(2).sort_values()
# print('Average of GDP per capita in different regions are :', data2)
print('The region which has the highest GDP per capita is :')
data2.nlargest(1)
```

```python
# b. International trade: Imports (million US$)
data3 = tdata.groupby('Region')['International trade: Imports (million US$)']. mean().round(2).sort_values()
data3.nlargest(1)
```

```python
# 5. Find out which region has more consistent with respect to GDP per capita (current US$)

data4= tdata.groupby('Region')['GDP per capita (current US$)'].std().round(3).sort_values()
print('The regions with respect to GDP per capita : \n', data4)

# Region which is more consistent w.r.t GDP per capita
print('\n \n The region which is more consistent with respect to GDP per Capita : \n',data4.nsmallest(1))

# Graph plotting Standard Deviation of GDP per Capita
fig=plt.subplots(figsize=(12,4))
plt.xticks(rotation=90)
sns.lineplot(data=data4, markers= True, color='teal').set(title='Standard Deviation of GDP per Capita')
plt.show()
print('From the graph and table,it can be observed that the Western African region has lesser standard devation, so it is more consistent (with respect to GDP per capita) followed by Northern African region.')
```
