```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
tdata=pd.read_csv("C:/Users/user/Desktop/Data Science/ACTIVITY/Telco-Customer-Churn.xlsx - Telco-Customer-Churn.csv")
tdata
```

```python
# Compare churn count with respect to gender
plt.figure(figsize=(10, 5))
sns.countplot(tdata['Churn'])
plt.title('Churn count with respect to Gender')
plt.xlabel('Churn')
plt.ylabel('Gender')
plt.show()
```

```python
#Find out how many female senior citizens there in the dataset
bar=np.arange(1)
Fem= tdata[(tdata['SeniorCitizen']==1) & (tdata['gender']=='Female')]
Fem1 = (Fem['SeniorCitizen'].sum())
print('Count of Female Senior Citizen : ', Fem1)
Mal= tdata[(tdata['SeniorCitizen']==1) & (tdata['gender']=='Male')]
Mal1 = (Mal['SeniorCitizen'].sum())
print('Count of Male Senior Citizen : ', Mal1)
plt.bar(bar,Fem1,width=0.50,color='pink', label='Female')
plt.bar(bar+0.50, Mal1,width=0.50,color='blue', label='Male')
plt.legend()
plt.xticks(bar+0.25,['SeniorCitizen'])
plt.ylabel('Females')
plt.title('Female Senior Citizens')
plt.show()
print('There are 568 Female Senior Citizens in the dataset.')
```

```python
# Compare 'tenure' with 'Total Charges'
sns.lineplot(data=tdata,x='tenure',y='TotalCharges', color='teal').set(title='Tenure vs Total Charges')
plt.xlabel('Tenure')
plt.ylabel('Total Charges')
plt.show()
print('From the graph, it can be observed that, as tenure is increasing, total charges also increasing.')
```
